## Helper functions

```python
def plot_img(img):
    print(type(img))
    plt.figure()
    plt.imshow(img)
    plt.show()

# convert bounding shape to bounding box
def bounding_box(points):
  points = points.astype(np.int16)
  x_coordinates, y_coordinates = zip(*points)
  return [(min(x_coordinates), min(y_coordinates)), (max(x_coordinates), max(y_coordinates))]
```

## Cropping images

```python
from PIL import Image


path = '/content/result_'
dict_values = {}
i = 0

files_docs = ['photos/'] #, 'CNH_Aberta/'  'CNH_Frente/', 'CNH_Verso/', 'CPF_Frente/', 'CPF_Verso/', 'RG_Aberta/', 'RG_Frente/', 'RG_Verso/']

for doc in files_docs:
  print(doc)

  txt_files = [f for f in os.listdir(path + doc) if f.endswith('.txt')]
  dict_values[doc] = {}

  txt_files = sorted(txt_files)

  for filename_txt in txt_files:
    new_filename = filename_txt.replace('.txt', '.jpg')
    new_filename = new_filename.replace('res_', '')
    print('file: ', new_filename)
    dict_values[doc][str(new_filename)] = {}

    crop_gray = ''
    crop_default = ''
    crop_threshold = ''
    i = 0
    with open(path + doc + filename_txt, 'r') as arquivo:
      for linha in arquivo:
          i = i + 1
          numeros = linha.split(",")
          box_coords = [int(x) for x in numeros]

          # Extraindo as coordenadas
          points = np.array(box_coords, dtype=np.int16)
          points = points.reshape(-1, 2)
          box_coords = bounding_box(points)
          box_coords = np.array(box_coords).reshape((-1, 2))
          box_coords = box_coords[np.argsort(box_coords[:, 1])]

          top_left = box_coords[0]
          bottom_right = box_coords[-1]
          w = bottom_right[0] - top_left[0]
          h = bottom_right[1] - top_left[1]
          x, y = top_left[0], top_left[1]

          image = cv2.imread('photos/' + new_filename)
          plt.imshow(image)

          # Aplicando os filtros
          image_full_gray =  cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
          text = pytesseract.image_to_string(image_full_gray , lang='por')
          clean_text = text.strip().replace("\n", " ").replace("\r", "")
          dict_values[doc][new_filename]['full_image'] = clean_text

          # Recortando a imagem
          crop_image = image[y-5:y+h+5, x-5:x+w+5]
          if crop_image.size != 0:
            crop_image = cv2.copyMakeBorder(
                src=crop_image,
                top=10,
                bottom=10,
                left=10,
                right=10,
                borderType=cv2.BORDER_CONSTANT,
                value=(255, 255, 255),
            )

          image_pil = Image.fromarray(crop_image)
          filename_crop = filename_txt.replace('.txt', '')
          if i < 10:
            ide = str(0) + str(i)
          else:
            ide = i
          image_pil.save(f"/content/crop/{doc}{filename_crop}_{ide}.jpg")  # Save as JPEG format

```
